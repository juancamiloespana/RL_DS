```python
from egreedy_decay_JCE import optimize_lotka_volterra
```

```python

# Define initial parameters and bounds
initial_params = [0.002, 0.04, 0.1, 0.0025]
param_bounds = [[0.0015, 0.0025], [0.03, 0.05], [0.05, 0.15], [0.002, 0.003]]

# Run the optimization
optimized_params, log= optimize_lotka_volterra(initial_params, param_bounds, runs=1)

```

```python
from tqdm import tqdm

# Create an empty list to store the results
results = []

# Loop for 30 optimization runs
for i in tqdm(range(30)):
  start_time = time.time()
  optimized_params, log = optimize_lotka_volterra(initial_params, param_bounds, runs=1000)
  # Calculate the average of the log (assuming 'log' is a list of values)
  end_time = time.time()
  iteration_time = end_time - start_time
  print(f"Iteration {i+1} completed in {iteration_time:.4f} seconds")

  # Append results for this iteration to the list
  for step, val in enumerate(log):
      results.append([i + 1, step + 1, val])

# Create a pandas DataFrame from the results
df = pd.DataFrame(results, columns=['Iteration', 'Step', 'avg_wolvw'])

# Now 'df' contains the results of all 30 optimization runs.
# You can save this to a CSV file or further process it as needed.
# Example: Save to a CSV
df.to_csv("optimization_results.csv", index=False)


df
```

```python
# prompt: Using dataframe df: create a graph with the mean  of avg_wolf group by step with the percentile 97.5 and 2.5 for all values in iteration gru¿oped by step

import pandas as pd
import altair as alt

# Group data by 'Step' and 'Iteration' and calculate the mean of 'avg_wolvw'
grouped_df = df.groupby(['Step'])['avg_wolvw'].mean().reset_index()

# Calculate the 2.5th and 97.5th percentiles of 'avg_wolvw' for each 'Step'
percentiles = df.groupby('Step')['avg_wolvw'].quantile([0.025, 0.975]).unstack()
percentiles = percentiles.reset_index()
percentiles.columns = ['Step', 'lower', 'upper']

# Merge the percentiles with the grouped data
merged_df = pd.merge(grouped_df, percentiles, on='Step')

```

```python
import matplotlib.pyplot as plt
# Create the Altair chart
plt.plot(merged_df['Step'] , merged_df["avg_wolvw"], color="black")
plt.fill_between(merged_df['Step'] , merged_df ["lower"], merged_df ["upper"], alpha=0.6, color="red")

plt.xlabel("Corrida")
plt.ylabel("Retorno promedio (y)")
plt.title("GRÁFICO RENDIMIENTO IC 95%")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

# AGRUPAR DATOS Y CALCULAR LOS DATOS PARA EL IC
df_agrupado = df_decay.groupby(["epsilon", "fac", "run"])["y"]
mean_y = df_agrupado.mean()
std_y = df_agrupado.std()
n = df_agrupado.count()

# CALCULO DEL IC 95%

# Compute percentiles
lower = df_agrupado.quantile(0.025)
upper = df_agrupado.quantile(0.975)


# GRÁFICO PRINCIPAL
df_summary = pd.DataFrame({
    "mean_y": mean_y,
    "upper": upper,
    "lower": lower
}).reset_index()

plt.figure(figsize=(10,6))

combinaciones = df_summary.groupby(["epsilon", "fac"])
colores_disponibles = plt.rcParams['axes.prop_cycle'].by_key()['color']  # Lista de colores de Matplotlib
colores = {}  # Diccionario para almacenar colores de cada serie

for (idx, ((nivel_e, nivel_fac), data)) in enumerate(combinaciones):
    color = colores_disponibles[idx % len(colores_disponibles)]  # Selecciona un color cíclicamente
    colores[(nivel_e, nivel_fac)] = color  # Guarda el color para usarlo después
    plt.plot(data["run"], data["mean_y"], label=f"E={nivel_e}, Fac={nivel_fac}", color=color)
    plt.fill_between(data["run"], data["lower"], data["upper"], alpha=0.2, color=color)


plt.xlabel("Run")
plt.ylabel("Avg reward")
plt.title("GRÁFICO RENDIMIENTO IC 95%")
max_run = df_summary["run"].max()
num_xticks = int(max_run / 100)
plt.xticks(np.linspace(1, max_run, num=num_xticks, dtype=int))
num_yticks = 10
y_min, y_max = df_summary["mean_y"].min(), df_summary["mean_y"].max()
plt.yticks(np.linspace(y_min, y_max, num=num_yticks))
plt.legend(title="Exploración, Paso", loc="upper left", fontsize=10, frameon=True, framealpha=0.3)
plt.grid(True)
plt.show()

```
